## **The binomial distribution applies when you have:**

1. A fixed number of trials (say, 10 coin flips).
2. Each trial has only two outcomes (success/failure, heads/tails).
3. The probability of success is the same each time.
4. Trials are independent (one coin flip doesn’t affect another).
If all those boxes are ticked, the binomial is your tool.



### **Exercise**
A class has 30 students, and each student has a 70% chance of passing an exam. What is the probability that at least 25 students pass?

```python
import math

p = 0.7 # probability of success
k = 25 # starting successes
n = 30 # total number of students

prob = 0.0

# binomial problem (pass or not pass)
for i in range(k, n+1):
    prob += math.comb(n, i) * p**i * (1-p)**(n-i)

prob = sum(math.comb(n, i) * (p ** i) * ((1 - p) ** (n - i)) for i in range(k, n + 1))

print(prob)
```

## **Combinations vs Permutations**
Combinations are the number of ways we can arrange items out of a mix without accounting for items order, and assuming that all the items are unique in the mix.

Permutations instead account for the order, and thus the formula changes.

The difference in computation is that permutations don't divide by k!

### **Exercise**
How many ways of making tris exist out of 20 people? --> combinations

And considering also order? --> permutations

```python
import math

n = 20 # total number of people
k = 3 # tris

combs = math.comb(20, 3)
print("there are", combs, "ways in which you can arrange people without considering order")

perms = math.perm(20, 3)
print("there are", perms, "ways in which you can arrange people considering order")
```

### **Exercise**
You roll a fair six-sided die 10 times. What is the probability of rolling a "6" exactly 3 times?

```python
# probability of rolling a "6" = 1/6
p = 1/6

n = 10 # you roll ten times in total
k = 3 # you roll three times six

prob = math.comb(10, 3) * p**k * (1-p)**(n-k)
print(prob)
```

## **Plotting now the binomial distribution**
Since there is the combination computation, its not possible to put the function in the continuous field. This distribution is discrete.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

n = 10

xs = list(np.arange(20)) # x axis
ys = [math.comb(n, k) * (p**k) * ((1-p)**(n-k)) for k in xs] # y axis

plt.plot(xs, ys)
```

### **Exercise**
“How many ways can five people sit around a lunch table?”

```python
n = 5 # total number of people

# if the people where in a line, the answer would be:
print('if people put in line, the answer is', math.factorial(n))

print('--------------')

# if the table instead is a circle, there is rotation that doesn't change the configuration, thus we have to decrease the count by 5 possible rotations
print('if the table is rounded, the answer is', math.factorial(n)/5)
print('which can also be seen as:', math.factorial(n-1))
```
