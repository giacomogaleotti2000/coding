## **PyTorch Basics**

```python
import torch

A = torch.rand(3,4)
B = torch.rand(4,1)
```

```python
A = A.to('cuda')
B = B.to('cuda')
```

```python
B
```

```python
A = torch.rand(3,4, requires_grad=True)
B = torch.rand(4,1, requires_grad=True)

C = A @ B
C
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = (x**C)

# visualizing the computation graph
from torchviz import make_dot
y = y.sum()
make_dot(y, params={'x':x, 'C':C}, show_attrs=True)
```

```python
y.backward()
```

```python
print(x.grad)
```
