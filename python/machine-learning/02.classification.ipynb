## **Classification**
The main difference with linear regression is that due to the need of changing the regression function from linear to logistic, also the cost function changes, from being the least squares optimization to logarithmic

```python
import numpy as np
import matplotlib.pyplot as plt

Xs = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n) --> m = samples numer (6) and n = features number (2)
ys = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1)                             #(m,1)

m = len(Xs)
print(f"{m} samples in the data")

print("Xs shape:", Xs.shape)
print("ys shape:", ys.shape)

plt.scatter(Xs[:,0], Xs[:,1], c=ys, cmap="viridis")
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# f(x) = 1 / [ 1 + e^-(W @ Xs + b) ]

# loss_func(f(x), y) = - 1/m sum [ y * log(f(x)) - (1 - y) * log(1-f(x)) ]

# initializing the weights matrix
W = np.array([0,0]).reshape(-1,1) # (2, 1)
b = 0

# learning rate
lr = 0.1

# track stats
losses = []

for _ in range(100000):

    Z = Xs @ W + b # (6, 2) @ (2, 1) = (6, 1)
    F = sigmoid(Z) # (6, 1)

    Loss = - 1/m * np.sum(ys * np.log(F) + (1 - ys) * np.log(1 - F))
    losses.append(Loss)

    dLdW = 1/m * Xs.T @ (F - ys) # (2, 6) @ (6, 1) = (2, 1) --> which is the dimension of W --> cool !
    dLdb = 1/m * np.sum(F - ys)

    W = W - lr * dLdW
    b = b - lr * dLdb

print("final W:", W)
print("final b:", b)
plt.plot(losses)
plt.show()
print(f"final loss: {losses[-1]:.6f}")
```
