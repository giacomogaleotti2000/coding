```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

reg = linear_model.LinearRegression()
X = np.array([0.9, 2, 4])
X = X.reshape(-1, 1)
y = np.array([0, 1, 2])
reg.fit(X, y)

plt.scatter(X, y, color='black')
plt.plot(X, reg.predict(X), color='blue', linewidth=1)
```
